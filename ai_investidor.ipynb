```python
#modificando o separador decimal de virgula para ponto e mudando o tipo para float
di['Taxa SELIC'] = di['Taxa SELIC'].str.replace(',','.').astype(float)
```

```python
di
```

```python
#mudando o formato da data
di['Data'] =  pd.to_datetime(di['Data'],format='%d/%m/%Y')
```

```python
#atribuindo a data como index
di.index = di['Data']
```

```python
di.drop(columns='Data', inplace = True)
```

#### Calculando rendimento total taxa DI

```python
#calculando do rendimento total da taxa di atraves do produto dos fatores diarios
rendimento_total_di = 1 + ((di['Fator Diário']).prod() - 1) 
rendimento_total_di
```

```python
#rendimento em reais
rendimento_di_real = 10000 * rendimento_total_di
rendimento_di_real
```

```python
#valor inicial investido
valor_inicial = 10000
```

```python
#calculo do total acumulado
di['valor_acumulado'] = valor_inicial * di['Fator Diário'].cumprod()
di
```

#### Normalizando deepseek

```python
#aplicando o mesmo procedimento para as carteiras
#normalizacao
cotacoes_deepseek_normalizado = cotacoes_deepseek/cotacoes_deepseek.iloc[0]
#atribuicao da data como indice
cotacoes_deepseek_normalizado['Date'] = cotacoes_deepseek_normalizado.index
#df com os ativos e as suas participacoes
contribuicao_deepseek = deepseek[['ticker','participacao']]
```

```python
#pivotamento do df 
deepseek_melted_normalizado = pd.melt(cotacoes_deepseek_normalizado, id_vars=['Date'], var_name='Ativo', value_name='Preco')
#merge das contribuicoes com as cotacoes
retorno_carteira_deepseek = deepseek_melted_normalizado.merge(contribuicao_deepseek, how = 'inner', left_on='Ativo', right_on = 'ticker')
#calculo do retorno por ativo
retorno_carteira_deepseek['contribuicao'] = retorno_carteira_deepseek['Preco'] * retorno_carteira_deepseek['participacao']
#groupby por data e contribuicao para obter o valor diario da carteira
retorno_carteira_deepseek = retorno_carteira_deepseek.groupby('Date')['contribuicao'].sum().reset_index()
#renomeando coluna para melhor identificar a feature
retorno_carteira_deepseek.rename(columns={'contribuicao': 'valor_carteira'}, inplace=True)
#valor em reais
retorno_carteira_deepseek['valor_carteira_reais'] = retorno_carteira_deepseek['valor_carteira'] * 10000
retorno_carteira_deepseek
```
